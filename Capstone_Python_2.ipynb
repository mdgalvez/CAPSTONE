```
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import numpy as np
from tabulate import tabulate
%matplotlib inline




def filter_dataframe(df,year,peformer,sortBy):
  df_filter = df[(df['Discharge Year'] == int(year)) & (df['3-Hour Bundle Quintile of Performance'] == peformer) ].sort_values(by='Met 3-Hour Bundle (%)',ascending=sortBy) 

  return df_filter

performer_choices = {'1': '1-Lowest','2':'2-Low','3':'3-Middle','4':'4-High','5':'5-Highest'}
sort_choices = {'1': True,'2':False}
#remove missing values
missing_values = [".", ",","na"]
data = pd.read_csv('https://raw.githubusercontent.com/mdgalvez/CAPSTONE/master/2015%20to%202016%20with%20Zip%20Code.csv', na_values = missing_values)

data.drop_duplicates(keep = 'first',inplace = True)

data = data[['Discharge Year', 'PFI','Facility Name','Location','Zip Code','Met 3-Hour Bundle (%)','3-Hour Bundle Quintile of Performance','Risk Adujsted Mortality Denominator','Observed Deaths','Risk Adjusted Mortality Rate (RAMR)']]

while True:
  discharge_year = input("Enter Discharge Year: ")
 
  print("Now please enter your choice for High/Low Peformer . \n 1-Lowest \n 2-Low \n 3-Middle \n 4-High \n 5-Highest")
  performer_choice = input("Enter Value: ")
  performer_value = performer_choices[performer_choice]
 
  print("Sort By Met 3-Hour Bundle (%) . \n 1-Ascending \n 2-Descending")
  sort_choice = input("Enter Value:")
  sort_value = sort_choices[sort_choice]
  

  df_filter = filter_dataframe(data,discharge_year,performer_value,sort_value)
  print(tabulate(df_filter, headers='keys', tablefmt='psql'))



  print("\nDo you want to continue? (Y/N)")
  ans = input()
      # if user input n or N then condition is True
  if ans == 'n' or ans == 'N':
    break




















 
```
